#### Funo Acha Framework

```python
def achaFrameWork(resumo, palavras=[]):
    status = False

    if(type(resumo) == str):
        resumo = resumo.lower()
    for palavra in palavras + ['framework', 'model']:
        if(type(resumo) == str):
            if(palavra in resumo):
                status = True     
    return status
    
    
```

#### Importao do Excel

```python
df = pd.read_excel('todos.xlsx', sheet_name='both')
```

#### Criao do Conjunto de palavras e Funes de Tratamento

```python
def trataPalavras(x):
    return x.lower().replace('(', '').replace(')', '')
```

```python
# foi feito uma transformao nas palavras para separar os termos das abreviaes destes
palavras = [' '.join(trataPalavras(x).split(' ')[1:]) for x in palavras_professor] + [' ' +(trataPalavras(x).split(' ')[0] + ' ') for x in palavras_professor]
```

#### Palavras Utilzadas para Anlise

```python
palavras[1:10]
```

```python
resultado = df[df['AB'].apply(lambda x: achaFrameWork(x, palavras)) == True]
```
